# REPLACE

```python
# replace col ---> Installs

df['Installs']= df['Installs'].replace("[, +]","", regex=True).astype(int)
df
```

```python
# size-----> 10M, 6K, varies with device
```

```python


df['Size']= df['Size'].replace("Varies with device","2.3", regex=True)
df
```

```python
df['Size'][679]
```

```python
df['Size']= df['Size'].replace("[M k ,]","", regex=True).astype(float)
df
```
